```python
import rsa

# Generate 2048-bit RSA key pair
(public_key, private_key) = rsa.newkeys(2048)

# Save them as files
with open("public.pem", "wb") as f:
    f.write(public_key.save_pkcs1())

with open("private.pem", "wb") as f:
    f.write(private_key.save_pkcs1())
```

```python
import rsa

# 1Ô∏è‚É£ Generate new keys (only once ‚Äî you can skip this later)
(public_key, private_key) = rsa.newkeys(2048)

# 2Ô∏è‚É£ Save keys to files
with open("public.pem", "wb") as f:
    f.write(public_key.save_pkcs1())
with open("private.pem", "wb") as f:
    f.write(private_key.save_pkcs1())

# 3Ô∏è‚É£ Encrypt a message using public key
message = "Hello Sultan!".encode('utf-8')
with open("public.pem", "rb") as f:
    pubkey = rsa.PublicKey.load_pkcs1(f.read())

ciphertext = rsa.encrypt(message, pubkey)
print("üîê Encrypted message (bytes):", ciphertext)

# 4Ô∏è‚É£ Decrypt it using private key
with open("private.pem", "rb") as f:
    privkey = rsa.PrivateKey.load_pkcs1(f.read())

plaintext = rsa.decrypt(ciphertext, privkey)
print("üí¨ Decrypted message:", plaintext.decode('utf-8'))
```
